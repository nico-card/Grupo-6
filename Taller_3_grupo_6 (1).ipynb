```python
import numpy as npy  # primero debemos importar la libreria que en este caso es numpy
from skimage import data # Ahora importamos el submodulo de la biblioteca skimage
import matplotlib.pyplot as plt # importamos el paquete matplotlib.pyplot  
c=[] # declaramos una matriz vacia
for a in range(0,50): #Establecemos un bucle for para a de 0 a 50 (Ancho eje x)
  h=[] # Establecemos una linsta en donde guardaremos los datos
  for b in range (0,50): #Establecemos un bucle for para a de 0 a 50 ( Alto eje y)
    if b<a: # ponemos una condicion de si b es mayor que a entonces...
     p=[0,0,0] #designamos 0 es decir el color negro a todos los valores de p
    else: #entonces si no se cumple la condicion anterior entonces...
     p=[255,255,255] # designamos 255 (el cual es el color blanco) a todos los valores del pixel
    h.append(p) #al final de cada ciclo del for anidado a単adimos la viarable p la cual es de pixel a la lista 
  c.append(h)#al final de cada ciclo del primer for o for "principal" se a単ade la lista y todos los datos a単adidos a la matriz h
plt.imshow(c) #se muestra en la pantalla la imagen creada

```
